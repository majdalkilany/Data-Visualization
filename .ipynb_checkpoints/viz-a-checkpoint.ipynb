```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid= pd.read_csv('covid.csv')
sns.set()
```

```python
covid.head(5)
```

```python
covid_20 = covid[['Confirmed' ,'Country/Region']].sort_values('Confirmed', ascending=False).nlargest(10 , 'Confirmed')
```

```python
covid_20
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(
    data=covid_20,
    x='Country/Region',
    y='Confirmed',
    color='red')
```

 ### the most efected country is us 

```python
covid_20  = covid_20 = covid[['Confirmed' ,'Country/Region', 'Deaths']].sort_values('Confirmed', ascending=False).nlargest(20 , 'Confirmed')
```

```python
covid_20
```

```python
plt.figure(figsize=(20,5))
graph = sns.barplot(x="Country/Region", y="Deaths", data=covid_20, ci=None)
```
